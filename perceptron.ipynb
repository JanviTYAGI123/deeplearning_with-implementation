```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# AND gate dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])  # AND labels

# Train Perceptron
clf = Perceptron(max_iter=1000, eta0=0.1, random_state=42)
clf.fit(X, y)

print("Trained weights:", clf.coef_)
print("Trained bias:", clf.intercept_)

# Plot dataset
plt.figure(figsize=(6,6))
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i,0], X[i,1], color='red', marker='o', s=100, label="Class 0" if i==0 else "")
    else:
        plt.scatter(X[i,0], X[i,1], color='blue', marker='x', s=100, label="Class 1" if i==3 else "")

# Decision boundary
x_values = np.linspace(-0.5, 1.5, 100)
# Equation: w1*x1 + w2*x2 + b = 0
y_values = -(clf.coef_[0][0]*x_values + clf.intercept_[0]) / (clf.coef_[0][1] + 1e-6)
plt.plot(x_values, y_values, color="green", linestyle="--", label="Decision Boundary")

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron (sklearn) Decision Boundary - AND Gate")
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.legend()
plt.grid(True)
plt.show()
```
